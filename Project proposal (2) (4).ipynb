We had to do some additional wranging to our data, to make our visualizations possible (because of the time-nature of the predictors we are measuring). Our exploratory visualizations begin by converting  the original_start_time and original_end_time to proper date-time objects: hour_of_day and day_of_week and month. The original_start_time and original_end_time columns store timestamps in UNIX milliseconds (e.g., 1638316800000), so we had to convert them by dividing by 1000 and using as.POSIXct(). Calculating the length of each session was obtained by subtracting start_time from end_time. We then removed the rows with missing timestamps as mentioned earlier after checking for them, then created a histogram for the distribution of gaming sessions throughout the day (24hrs), a bar plots for the 7 days of the week, and a bar plot for the distribution of start times throughout the availible months. Additionally, we added a histogram to display how long gaming sessions typically last, helping us understand if players prefer short periods or longer periods of play (in progress, couldnt make it work in time for the deadline). The visualizations reveal clear patterns in player activity: peak hours (e.g., late afternoon/evening), higher activity on weekends, and potential seasonal trends (e.g., more play in summer). These insights directly support our research question by showing that time-based features (hour, day, month) are strong predictors of player activity.

```r
sessions_data <- sessions_data %>%
  mutate(
    start_time = as.POSIXct(original_start_time / 1000, origin = "1970-01-01", tz = "UTC"),
    end_time = as.POSIXct(original_end_time / 1000, origin = "1970-01-01", tz = "UTC")
  )

sessions_data <- sessions_data %>%
  mutate(
    hour_of_day = hour(start_time), # Extract hour of the day
    day_of_week = wday(start_time, label = TRUE, abbr = FALSE), # Extract day of the week
    month = month(start_time, label = TRUE, abbr = FALSE) # Extract month
  )

sessions_data <- sessions_data %>%
  mutate(session_duration = as.numeric(difftime(end_time, start_time, units = "mins")))

sum(is.na(sessions_data$original_start_time))
sum(is.na(sessions_data$original_end_time))

invalid_durations <- sessions_data %>%
  filter(end_time < start_time)
nrow(invalid_durations)

sessions_data_clean <- sessions_data %>%
  filter(!is.na(original_start_time)) %>%
  filter(!is.na(original_end_time)) %>%
  filter(end_time >= start_time)

ggplot(sessions_data_clean, aes(x = hour_of_day)) +
  geom_histogram(binwidth = 1, fill = "skyblue", color = "black") +
  labs(
    title = "Distribution of Session Start Times by Hour of the Day",
    x = "Hour of the Day",
    y = "Number of Sessions"
  ) +
  theme_minimal()

ggplot(sessions_data_clean, aes(x = day_of_week)) +
  geom_bar(fill = "lightgreen", color = "black") +
  labs(
    title = "Distribution of Session Start Times by Day of the Week",
    x = "Day of the Week",
    y = "Number of Sessions"
  ) +
  theme_minimal()

ggplot(sessions_data_clean, aes(x = month)) +
  geom_bar(fill = "orange", color = "black") +
  labs(
    title = "Distribution of Session Start Times by Month",
    x = "Month",
    y = "Number of Sessions"
  ) +
  theme_minimal()

```
